## NumPy Basics

NumPy is one of the most important foundations in the Python data-science stack. It provides data storage and computation capabilities, and many other data-science libraries build on top of it. NumPy's core type is `ndarray`, which represents arrays of arbitrary dimension. Compared with Python `list`, it has the following advantages:

1. Better performance: it can take advantage of hardware parallelism and cache optimization, producing orders-of-magnitude performance gains over `list` in data processing.
2. More powerful functionality: `ndarray` provides rich operations and methods for data processing, and NumPy also includes many functions built specifically for array operations.
3. Vectorized operations: functions in NumPy and methods on `ndarray` act on the whole array, so explicit loops are often unnecessary and the code becomes simpler and cleaner.

```python
# Interview question: what is an iterator in Python, and how is it related to a generator?
# An iterator is an object that implements the iterator protocol. In Python, that protocol is defined by the two magic methods `__iter__` and `__next__`.
# We can retrieve data from an iterator with the `next` function or a `for-in` loop.
# Writing iterators directly can be cumbersome, so in Python generators are often used to simplify iterator syntax.


def fib(count):
    a, b = 0, 1
    for _ in range(count):
        a, b = b, a + b
        yield a


gen = fib(50)
gen
```
