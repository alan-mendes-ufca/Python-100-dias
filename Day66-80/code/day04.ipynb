## Uma introdução prática aos pandas 1

pandas é uma biblioteca Python de código aberto que oferece suporte ao fluxo de trabalho completo de análise de dados. Wes McKinney começou a desenvolvê-lo em 2008 com o objetivo de fornecer ferramentas práticas para análise e processamento de dados em larga escala. O pandas envolve operações como carregamento, remodelagem, limpeza, dinamização e apresentação de dados e fornece três tipos de dados principais:
1. `Series`: uma estrutura de dados rotulada unidimensional. Ao contrário de uma matriz unidimensional simples, cada valor possui um índice e oferece métodos de processamento de dados mais ricos do que `ndarray`.
2. `DataFrame`: uma estrutura de dados bidimensional semelhante a uma tabela. Comparado com uma matriz bidimensional, um `DataFrame` possui índices de linha e coluna e oferece mais de 100 métodos para processamento de dados.
3. `Index`: a estrutura de indexação usada por `Series` e `DataFrame`.

```python
# Detectar valores discrepantes
# Regra numérica (dados < Q1 - 1,5 * IQR ou dados > Q3 + 1,5 * IQR)


def find_outliers_by_iqr(data, whis=1.5):
    q1, q3 = np.quantile(data, [0.25, 0.75])
    iqr = q3 - q1
    return data[(data < q1 - whis * iqr) | (data > q3 + whis * iqr)]
```

```python
temp = np.random.normal(80, 8, 50).round(0)
temp = np.append(temp, [120, 160, 200, 40, 20, -50])
temp
```

```python
find_outliers_by_iqr(temp)
```

```python
find_outliers_by_iqr(temp, whis=3)
```

```python
# método z-score (regra de três sigma -> regra 68-95-99,7)


def find_outliers_by_zscore(data, mul=3):
    mu, sigma = np.mean(data), np.std(data)
    zscore = (data - mu) / sigma
    return data[np.abs(zscore) > mul]
```

```python
find_outliers_by_zscore(temp)
```

```python
find_outliers_by_zscore(temp, mul=2)
```

#### Caso 2: pré-processamento de dados de pontos de residência em Pequim

1. Carregue os dados
2. Processamento de data e hora
3. Binning por faixa etária
4. Normalize os pontos de residência

```python
settle_df = pd.read_csv('res/2023-beijing-residence-points-data.csv', index_col='announcement_id')
settle_df.head(5)
```

```python
settle_df.info()
```

```python
# Converter strings em datas
settle_df['birth_year_month'] = pd.to_datetime(settle_df['birth_year_month'])
settle_df.info()
```

```python
# Converta datas de nascimento em idades
settle_df['age'] = (pd.to_datetime('2023-01-01') - settle_df.birth_year_month).dt.days // 365
settle_df.head(5)
```

```python
# Divida as idades em faixas etárias - armazenamento - baldes
settle_df['age_group'] = pd.cut(
    settle_df.age,
    bins=np.arange(35, 61, 5),
    labels=['35-39', '40-44', '45-49', '50-54', '55-59'],
    right=False
)
settle_df.head(5)
```

```python
# Conte a frequência de cada elemento
temp = settle_df.age_group.value_counts()
temp
```

```python
plt.cm.Greens(np.linspace(0.9, 0.1, 5))
```

```python
# Trace um gráfico de barras
temp.plot(
    kind='bar',      # tipo de gráfico
    figsize=(8, 4),  # tamanho do gráfico
    xlabel='',       # rótulo do eixo x
    ylabel='Count',  # rótulo do eixo y
    width=0.5,       # largura da barra
    hatch='//',      # padrão de hachura de barra
    color=plt.cm.Greens(np.linspace(0.9, 0.3, temp.size))  # valores de cores
)

for i in range(temp.size):
    # plt.text(x_coordenada, y_coordenada, rótulo_texto)
    plt.text(i, temp.iloc[i] + 30, temp.iloc[i], ha='center')

# Personalize os ticks do eixo x
plt.xticks(rotation=0)
plt.show()
```

```python
# agg - agregado
settle_df.points.agg(['mean', 'max', 'min', 'std', 'skew', 'kurt'])
```

Normalização linear: $$ x^{\prime} = \frac{x - x_{min}}{x_{max} - x_{min}} $$

```python
# Mapeie os valores dos pontos no intervalo 0-1
max_score, min_score = settle_df.points.agg(['max', 'min'])
max_score, min_score
```

```python
# map - aplica a função especificada a cada elemento de uma série
# aplicar - aplica a função especificada a cada elemento de uma série
settle_df['normalized_points'] = settle_df.points.map(lambda x: (x - min_score) / (max_score - min_score)).round(2)
settle_df
```

padronização do escore z: $$ x^{\prime} = \frac{x - \mu}{\sigma} $$

```python
mu, sigma = settle_df.points.agg(['mean', 'std'])
settle_df['zscore_score'] = settle_df.points.apply(lambda x: (x - mu) / sigma)
settle_df
```
