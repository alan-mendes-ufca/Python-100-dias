## A Hands-on Introduction to pandas 1

pandas is an open-source Python library that supports the full data-analysis workflow. Wes McKinney began developing it in 2008 with the goal of providing practical tools for large-scale data analysis and processing. pandas wraps operations such as data loading, reshaping, cleaning, pivoting, and presentation, and it provides three core data types:
1. `Series`: a one-dimensional labeled data structure. Unlike a plain one-dimensional array, every value has an index, and it offers richer data-processing methods than `ndarray`.
2. `DataFrame`: a two-dimensional table-like data structure. Compared with a two-dimensional array, a `DataFrame` has both row and column indexes and offers more than 100 methods for processing data.
3. `Index`: the indexing structure used by `Series` and `DataFrame`.

```python
# Detect outliers
# Numerical rule (data < Q1 - 1.5 * IQR or data > Q3 + 1.5 * IQR)


def find_outliers_by_iqr(data, whis=1.5):
    q1, q3 = np.quantile(data, [0.25, 0.75])
    iqr = q3 - q1
    return data[(data < q1 - whis * iqr) | (data > q3 + whis * iqr)]
```

```python
temp = np.random.normal(80, 8, 50).round(0)
temp = np.append(temp, [120, 160, 200, 40, 20, -50])
temp
```

```python
find_outliers_by_iqr(temp)
```

```python
find_outliers_by_iqr(temp, whis=3)
```

```python
# z-score method (three-sigma rule -> 68-95-99.7 rule)


def find_outliers_by_zscore(data, mul=3):
    mu, sigma = np.mean(data), np.std(data)
    zscore = (data - mu) / sigma
    return data[np.abs(zscore) > mul]
```

```python
find_outliers_by_zscore(temp)
```

```python
find_outliers_by_zscore(temp, mul=2)
```

```python
find_outliers_by_zscore(df6.direct_cost)
```

```python
# Delete rows based on the row indexes of outliers
df6.drop(index=find_outliers_by_zscore(df6.direct_cost).index)
```

```python
jobs_df.year.unique()
```

```python
jobs_df['edu'] = jobs_df.edu.replace(r'Technical Secondary School|High School', 'No Requirement', regex=True)
jobs_df['edu'] = jobs_df.edu.replace(r'Master|Doctorate', 'Graduate Degree', regex=True)
```

```python
jobs_df.edu.unique()
```

```python
jobs_df['job_name'] = jobs_df.job_name.str.lower()
jobs_df = jobs_df[jobs_df.job_name.str.contains('python|data|product|operations', regex=True)]
jobs_df.shape
```

#### Case 2: preprocessing Beijing residence-points data

1. Load the data
2. Date and time processing
3. Age-band binning
4. Normalize the residence points

```python
settle_df = pd.read_csv('res/2023-beijing-residence-points-data.csv', index_col='announcement_id')
settle_df.head(5)
```

```python
settle_df.info()
```

```python
# Convert strings to dates
settle_df['birth_year_month'] = pd.to_datetime(settle_df['birth_year_month'])
settle_df.info()
```

```python
# Convert birth dates into ages
settle_df['age'] = (pd.to_datetime('2023-01-01') - settle_df.birth_year_month).dt.days // 365
settle_df.head(5)
```

```python
# Divide ages into age bands - binning - buckets
settle_df['age_group'] = pd.cut(
    settle_df.age,
    bins=np.arange(35, 61, 5),
    labels=['35-39', '40-44', '45-49', '50-54', '55-59'],
    right=False
)
settle_df.head(5)
```

```python
# Count the frequency of each element
temp = settle_df.age_group.value_counts()
temp
```

```python
# agg - aggregate
settle_df.points.agg(['mean', 'max', 'min', 'std', 'skew', 'kurt'])
```

Linear normalization:
$$
x^{\prime} = \frac{x - x_{min}}{x_{max} - x_{min}}
$$

```python
# Map point values into the 0-1 range
max_score, min_score = settle_df.points.agg(['max', 'min'])
max_score, min_score
```

```python
# map - apply the specified function to each element in a Series
# apply - apply the specified function to each element in a Series
settle_df['normalized_points'] = settle_df.points.map(lambda x: (x - min_score) / (max_score - min_score)).round(2)
settle_df
```

z-score standardization:
$$
x^{\prime} = \frac{x - \mu}{\sigma}
$$

```python
mu, sigma = settle_df.points.agg(['mean', 'std'])
settle_df['zscore_score'] = settle_df.points.apply(lambda x: (x - mu) / sigma)
settle_df
```
