#### Vectorizing Regular Functions

```python
# Use the `vectorize` decorator to turn a regular function into a vectorized one
@np.vectorize
def fac(n):
    if n == 0:
        return 1
    return n * fac(n - 1)
```

### Broadcasting

If two arrays with different shapes need to participate in an operation, broadcasting must first make their shapes compatible.<br>
To use broadcasting so that two arrays can operate together, one of the following conditions must be met:

1. The trailing dimensions of the two arrays (the matching parts when reading `shape` from back to front) are the same.
2. The trailing dimensions differ, but one side is 1.

\boldsymbol{x} = \begin{bmatrix}
x_1 \\

$$

```python
m3 = np.arange(1, 10, dtype='f8').reshape(3, 3)
m3[-1, -1] = 8
m3
```

```python
# Calculate the rank of a matrix
np.linalg.matrix_rank(m3)
```

```python
# Inverse matrix - a singular matrix cannot be inverted
# LinAlgError: Singular matrix
np.linalg.inv(m3)
```

```python
# Condition for a unique solution: the rank of the coefficient matrix equals the rank of the augmented matrix, and both equal the number of unknowns.
# Rank: the number of linearly independent rows or columns.
# Linear dependence: if one vector can be obtained from others by linear transformation (scalar multiplication and addition), then they are linearly dependent.
```

```python
A = np.array([[1, 2, 1], [3, 7, 2], [2, 2, 1]])
b = np.array([8, 23, 9]).reshape(-1, 1)
```

```python
# Rank of the coefficient matrix
np.linalg.matrix_rank(A)
```

```python
# Rank of the augmented matrix
np.linalg.matrix_rank(np.hstack((A, b)))
```

```python
# Solve a system of linear equations
np.linalg.solve(A, b)
```

$$
A \cdot x = b
$$
$$
A^{-1} \cdot A \cdot x = A^{-1} \cdot b
$$
$$
I \cdot x = A^{-1} \cdot b
$$
$$
x = A^{-1} \cdot b
$$

```python
# Solve a system of linear equations using the inverse matrix
np.linalg.inv(A) @ b
```

```python
from scipy.ndimage import gaussian_filter, sobel

# Get the grayscale image
guido_image = plt.imread('res/guido.jpg')
gray_image = np.mean(guido_image, axis=2)

plt.figure(figsize=(12, 4))

# Grayscale image
plt.subplot(1, 4, 1)
plt.imshow(gray_image, cmap=plt.cm.gray)

# Blur and sharpen
plt.subplot(1, 4, 2)
blurred_image = gaussian_filter(gray_image, 3)
plt.imshow(blurred_image, cmap=plt.cm.gray)

plt.subplot(1, 4, 3)
filtered_image = gaussian_filter(blurred_image, 1)
sharpen_image = blurred_image + 32 * (blurred_image - filtered_image)
plt.imshow(sharpen_image, cmap=plt.cm.gray)

# Edge image
plt.subplot(1, 4, 4)
# Use the Sobel operator (weighted grayscale differences of neighboring points) for edge detection
edge_image = sobel(gray_image)
plt.imshow(edge_image, cmap=plt.cm.gray)

plt.show()
```

```python
# Goal: because monthly income and online-shopping expenses are strongly correlated, we can use monthly income to predict online-shopping expenses
# Method 1: given a monthly income, find the N closest data points and use their average to predict the corresponding online-shopping expenses
# KNN - k-nearest neighbors (find the k nearest neighbors and use their data to make a prediction)
import heapq


def predicate_by_knn(income, k=5):
    """KNN algorithm"""
    keys = heapq.nsmallest(k, history_data, key=lambda x: (x - income) ** 2)
    return np.mean([history_data[key] for key in keys]).round(2)
```

```python
predicate_by_knn(6800)
```

```python
predicate_by_knn(20000, k=3)
```

Regression model:
$$ Y = aX + b $$

Loss function:
$$ MSE = \frac{1} {N} \sum (\hat{y_i} - y_i)^2 $$

```python
# MSE - Mean Squared Error
def get_loss(a, b):
    """Loss function"""
    y_hat = a * x + b
    return np.mean((y_hat - y) ** 2)
```

```python
# Monte Carlo simulation
import random

min_loss = np.inf
ba, bb = None, None

for _ in range(10000):
    a = random.random() * 0.5 + 0.5
    b = random.random() * 1000 - 2000
    curr_loss = get_loss(a, b)
    if curr_loss < min_loss:
        min_loss = curr_loss
        ba, bb = a, b
        print(min_loss)
print(ba, bb)
```

```python
plt.scatter(x, y)
plt.plot(x, ba * x + bb, color='r', linewidth=4)
plt.show()
```

```python
def predicate_by_regression(income):
    return round(ba * income + bb, 2)
```

```python
predicate_by_regression(6800)
```

```python
predicate_by_regression(12800)
```

Substitute the regression model into the loss function:
$$ f(a, b) = \frac {1} {N} \sum_{i=1}^{N}(y_i - (ax_i + b))^2 $$

How can we make $f(a, b)$ reach its minimum?

Take the partial derivatives and set them equal to 0.
$$ \frac {\partial {f(a, b)}} {\partial {a}} = \frac {2} {N} \sum_{i=1}^{N}(-x_iy_i + x_i^2a + x_ib) = 0 $$ 
$$ \frac {\partial {f(a, b)}} {\partial {b}} = \frac {2} {N} \sum_{i=1}^{N}(-y_i + x_ia + b) = 0 $$

Solving gives:
$$a = \frac{\sum(x_{i} - \bar{x})(y_{i} - \bar{y})}{\sum(x_{i} - \bar{x})^{2}}$$
$$b = \bar{y} - a\bar{x}$$

```python
x_bar, y_bar = np.mean(x), np.mean(y)
```

```python
ba = np.dot((x - x_bar), (y - y_bar)) / np.sum((x - x_bar) ** 2)
bb = y_bar - ba * x_bar
ba, bb
```

```python
plt.scatter(x, y)
plt.plot(x, ba * x + bb, color='r', linewidth=4)
plt.show()
```
