## NoĂ§Ăµes bĂˇsicas de NumPy

NumPy Ă© uma das bases mais importantes da pilha de ciĂŞncia de dados Python. Ele fornece armazenamento de dados e recursos de computaĂ§ĂŁo, e muitas outras bibliotecas de ciĂŞncia de dados sĂŁo construĂ­das sobre ele. O tipo principal do NumPy Ă© `ndarray`, que representa matrizes de dimensĂŁo arbitrĂˇria. Comparado com Python `list`, tem as seguintes vantagens:

1. Melhor desempenho: pode aproveitar o paralelismo de hardware e a otimizaĂ§ĂŁo de cache, produzindo ganhos de desempenho de ordens de magnitude em relaĂ§ĂŁo ao `list` no processamento de dados.
2. Funcionalidade mais poderosa: `ndarray` fornece operaĂ§Ăµes e mĂ©todos avanĂ§ados para processamento de dados, e NumPy tambĂ©m inclui muitas funĂ§Ăµes criadas especificamente para operaĂ§Ăµes de array.
3. OperaĂ§Ăµes vetorizadas: funĂ§Ăµes em NumPy e mĂ©todos em `ndarray` atuam em todo o array, portanto, loops explĂ­citos sĂŁo frequentemente desnecessĂˇrios e o cĂłdigo se torna mais simples e limpo.

```python
# Pergunta da entrevista: o que Ă© um iterador em Python e como ele estĂˇ relacionado a um gerador?
# Um iterador Ă© um objeto que implementa o protocolo do iterador. Em Python, esse protocolo Ă© definido pelos dois mĂ©todos mĂˇgicos `__iter__` e `__next__`.
# Podemos recuperar dados de um iterador com a funĂ§ĂŁo `next` ou um loop `for-in`.
# Escrever iteradores diretamente pode ser complicado, portanto, em Python, geradores sĂŁo frequentemente usados â€‹â€‹para simplificar a sintaxe do iterador.


def fib(count):
    a, b = 0, 1
    for _ in range(count):
        a, b = b, a + b
        yield a


gen = fib(50)
gen
```
